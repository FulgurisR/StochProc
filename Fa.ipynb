```python
%pylab inline
from random import randrange
```

```python
def randRecurTree(N):
    '''N a csúcsok száma'''
    
    treeRand=[1,0] #kezdetben van 2 csúcs
                    #összekötve egymással
    i=2
    while i<N: #i=2,...,N-1
        treeRand.append(randrange(0,i-1)) #lista bővítése
                                            #sorsolt számmal
        i+=1
    
    return treeRand
```

```python
def fokszamolo(treeRand):
    
    fokszamok=[]
    i=0
    while i<N: #i=2,...,N-1
        j=0
        cnt=0
        while j<N: #j=2,...,N-1
            if treeRand[j]==i:
                cnt+=1
            j+=1
        fokszamok.append(cnt)    
        i+=1
        
    
    
    return fokszamok
```

#### 4 mérés $N=100$ esetére

```python
N=100
```

```python
tree100a=randRecurTree(N)
grid100=array([x for x in range(1,len(tree100a)+1)])
plot(grid100,tree100a,'.', label=r'$N= 100$')
grid()
xlabel('csúcsazonosító', fontsize=14)
ylabel('sorsolt csúcsazonosító', fontsize=14)
title('véletlen rekurzív fa ', fontsize=14)
legend(prop={'size': 14})
savefig('tree100a.png')
```
