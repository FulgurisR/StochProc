```python
%pylab inline
import random
import imageio
```

```python
def myMonteCarlo(N, betaJ):
    
    
    spins = ones(N) #tömb N db spin tárolására
                    #szélső helyzetből indulunk
    
    
    j =0
    while mean(spins)!=0: #amíg a mágnesezettség nem zérus
        mySpin=random.randint(0,49) #véletlesnszerűen kiválasztok egy spint
        if mySpin==0:
            deltaE = spins[mySpin]*(spins[49]+spins[mySpin+1]) #energiaváltozás
                                                #sorsolt és sorsolttal szomszédos
                                                #spinekből
        elif mySpin==49:
            deltaE = spins[mySpin]*(spins[mySpin-1]+spins[0]) #energiaváltozás
                                                #sorsolt és sorsolttal szomszédos  
                                                #spinekből
        else:
            deltaE = spins[mySpin]*(spins[mySpin-1]+spins[mySpin+1]) #energiaváltozás
                                                #sorsolt és sorsolttal szomszédos
                                                #spinekből
                                                                                                
        if deltaE<0:
            spins[mySpin]=-spins[mySpin] #forgatás
                
        elif deltaE==0:
            P=random.random() #véletlen valós sorsolása [0,1]-ből
            if P<0.5:
                spins[mySpin]=-spins[mySpin] #forgatás
            elif P>0.5:
                spins[mySpin]=spins[mySpin]
                
        else:
            P=random.random() #véletlen valós sorsolása [0,1]-ből
            if mySpin==0:
                expValue=exp(-2*betaJ*spins[mySpin]*(spins[49]+spins[mySpin+1]))
                
            elif mySpin==49:
                expValue=exp(-2*betaJ*spins[mySpin]*(spins[mySpin-1]+spins[0]))
            
            else:
                expValue=exp(-2*betaJ*spins[mySpin]*(spins[mySpin-1]+spins[mySpin+1]))
            
            if P<expValue:
                spins[mySpin]=-spins[mySpin] #forgatás
            else:    
                spins[mySpin]=spins[mySpin]     
        
        
        if mean(spins)==0:
            print('Relaxációs idő: ', j+1)
            
        j+=1
            
    return spins
```

#### 3 mérés $\beta J = 0.08$ esetére

```python
spins11=myMonteCarlo(50, 0.08)
gridSpins11=array([x for x in range(1,spins11.shape[0]+1)])
plot(gridSpins11,spins11,'.', label=r'$\beta J = 0.08$')
grid()
xlabel(r'$s_i$', fontsize=14)
ylabel('spin értéke', fontsize=14)
legend(prop={'size': 14})
savefig('spins11.png')
```
